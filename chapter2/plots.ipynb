```python
import json
import warnings
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
warnings.filterwarnings("ignore")
```

```python
iris = datasets.load_iris()

X = iris.data
y = iris.target

data = {
    "sepal_length": X[:, 0],
    "sepal_width": X[:, 1],
    "petal_length": X[:, 2],
    "petal_width": X[:, 3],
    "target": y,
}

target_names = {
    0: iris.target_names[0],
    1: iris.target_names[1],
    2: iris.target_names[2],
}

iris = pd.DataFrame(data)
iris["species"] = iris["target"].map(target_names)
iris.head()
```

```python
with open("data/style.json", "r") as f:
  rc = json.load(f)
sns.set_theme(rc=rc)

plt.figure(figsize=(8, 4))

ax = sns.lineplot(data=iris, x="sepal_length", y="sepal_width", hue="species")
ax.legend(title="Species", bbox_to_anchor=(1.05, 0.5), title_fontproperties={"weight": "bold"})
ax.set_title("Iris Dataset")
ax.set_xlabel("Sepal Length")
ax.set_ylabel("Sepal Width")

plt.show()
```

```python
with open("data/style.json", "r") as f:
  rc = json.load(f)
sns.set_theme(rc=rc)

plt.figure(figsize=(8, 4))

ax = sns.scatterplot(data=iris, x="sepal_length", y="sepal_width", hue="species")
ax.legend(title="Species", bbox_to_anchor=(1.05, 0.5), title_fontproperties={"weight": "bold"})
ax.set_title("Iris Dataset")
ax.set_xlabel("Sepal Length")
ax.set_ylabel("Sepal Width")

plt.show()
```

```python
with open("data/style.json", "r") as f:
  rc = json.load(f)
sns.set_theme(rc=rc)

plt.figure(figsize=(8, 4))

ax = sns.histplot(data=iris, x="sepal_length", kde=True)
ax.set_title("Iris Dataset")
ax.set_xlabel("Sepal Length")
ax.set_ylabel("Count")

plt.show()
```

```python
with open("data/style.json", "r") as f:
  rc = json.load(f)

mpl.rcParams.update(rc)
sns.set_theme(rc=rc)

plt.figure(figsize=(8, 4))

ax = sns.boxplot(data=iris, x="species", y="sepal_length")
ax.set_title("Iris Dataset")
ax.set_xlabel("Sepal Length")
ax.set_ylabel("Count")

plt.show()
```
